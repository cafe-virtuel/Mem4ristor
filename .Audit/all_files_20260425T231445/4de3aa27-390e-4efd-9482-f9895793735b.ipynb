```python
# Get all sections content - print key ones
def get_section(content, header):
    """Get content from a section header until the next header of same or higher level."""
    idx = content.find(header)
    if idx == -1:
        return None
    # find next ### or ##
    level = len(header) - len(header.lstrip('#'))
    rest = content[idx + len(header):]
    # find next header at same or higher level
    pat = re.compile(r'^#{1,' + str(level) + r'}\s+', re.MULTILINE)
    m = pat.search(rest)
    end = m.start() if m else len(rest)
    return content[idx:idx + len(header) + end]

key_sections = [
    "### 3sexies.",
    "### 3octies.",
    "### 3tertrigies.",
    "### 3quatervigies.",
    "### 3quinquevigies.",
    "### 3sexvigies.",
    "### 3septvigies.",
    "### 3octvigies.",
    "### 3novemvigies.",
    "### 3duotrigies.",
]
# find the actual headers
for line in content.split('\n'):
    for k in key_sections:
        if line.startswith(k):
            print(line)
            break
```

```python
# Compute group statistics
stats = sp50.groupby('point').agg({'H_continuous': ['mean', 'std', 'count'], 'H_cognitive_kimi': ['mean', 'std']})
print(stats)
# Effect size A vs B
import numpy as np
a = sp50[sp50.point=='A_dead_zone'].H_continuous.values
b = sp50[sp50.point=='B_noise_only'].H_continuous.values
c = sp50[sp50.point=='C_noise_cmos'].H_continuous.values
def cohen_d(x, y):
    pooled = np.sqrt(((len(x)-1)*x.var(ddof=1) + (len(y)-1)*y.var(ddof=1))/(len(x)+len(y)-2))
    return (y.mean() - x.mean())/pooled

print(f'A: n={len(a)} mean={a.mean():.3f} std={a.std(ddof=1):.3f}')
print(f'B: n={len(b)} mean={b.mean():.3f} std={b.std(ddof=1):.3f}')
print(f'C: n={len(c)} mean={c.mean():.3f} std={c.std(ddof=1):.3f}')
print(f'Cohen d (A→B) = {cohen_d(a,b):.3f}')
print(f'Cohen d (B→C) = {cohen_d(b,c):.3f}')
from scipy import stats as st
t, p = st.ttest_ind(a, b, equal_var=False)
print(f'Welch t A vs B: t={t:.2f}, p={p:.2e}')
t, p = st.ttest_ind(b, c, equal_var=False)
print(f'Welch t B vs C: t={t:.2f}, p={p:.2e}')
```
